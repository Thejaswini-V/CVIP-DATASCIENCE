# EXPLORATORY DATA ANALYSIS

## AVG. WATCHTIME

```python
time=df['Runtime']
sum=0
for i in time:
    i=i.replace("min","")
    sum=sum+int(i)
   
avg_time=sum/len(time)
avg_time
```

##  No. of movies in a year with rating>8.5

```python
high_ratings = df[df['IMDB_Rating'] > 8.5]
years_with_high_ratings = high_ratings.groupby('Released_Year').size().reset_index(name='Count')
```

```python
years_with_high_ratings
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Released_Year', y='Count', data=years_with_high_ratings)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Years with Ratings >8.5')
plt.xticks(rotation=90)
plt.show()
```

## NO. OF MOVIES IN A GENRE

```python
df['Genre'].value_counts()
```

```python
genre_counts = df['Genre'].value_counts()
genres = list(genre_counts.index)[:10]
counts = list(genre_counts.values)[:10]

plt.barh(y=genres, width=counts)
plt.show()
```

```python
top_directors = df['Director'].value_counts().head(10)
top_directors.plot(kind='bar')
plt.ylabel('Number of Movies')
plt.xlabel('Director')
plt.title('Top 10 Directors with the Most Movies')
plt.show()
```

## NO. OF MOVIES RELEASED PER YEAR

```python

movies_per_year = df['Released_Year'].value_counts().sort_index()
plt.plot(movies_per_year.index,movies_per_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies Released per Year')
x_ticks_step = 5
years = df['Released_Year'].unique()  
years_sorted = sorted(years) 
plt.xticks(range(0, len(years_sorted), x_ticks_step), years_sorted[::x_ticks_step], rotation=45)
plt.tight_layout()
plt.show()
```

## RELATION BETWEEN RUNTIME AND RATING

```python
plt.scatter(df['Runtime'], df['IMDB_Rating'])
plt.xlabel('Runtime')
plt.ylabel('Rating')
plt.title('Relationship between Runtime and Rating')
x_ticks_step = 50
Runtime_sorted=sorted(df['Runtime'])
plt.xticks(range(0, len(Runtime_sorted), x_ticks_step), Runtime_sorted[::x_ticks_step], rotation=45)
plt.tight_layout()
plt.xticks(rotation=45) 
plt.tight_layout() 
plt.show()
```

## RATING COUNTS

```python
imdb_ratings_counts = df['IMDB_Rating'].value_counts().sort_index()

plt.bar(imdb_ratings_counts.index, imdb_ratings_counts.values)
plt.xlabel('IMDB Rating')
plt.ylabel('Count')
plt.title('Distribution of IMDB Ratings')
plt.show()
```
